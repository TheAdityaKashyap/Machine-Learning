```python
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Our data
X = np.array([8, 10, 12]).reshape(-1, 1)  # Pizza diameters (independent variable)
Y = np.array([10, 13, 16])               # Pizza prices (dependent variable)

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Get the slope (m) and intercept (b)
m = model.coef_[0]  # Slope
b = model.intercept_  # Intercept
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")

# Predict price for a 20-inch pizza
X_new = np.array([20]).reshape(-1, 1)
Y_pred = model.predict(X_new)
print(f"Predicted price for a 20-inch pizza: ${Y_pred[0]:.2f}")

# Plot the data and the regression line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Diameter (inches)')
plt.ylabel('Price ($)')
plt.title('Pizza Price Prediction')
plt.legend()
plt.show()
```
